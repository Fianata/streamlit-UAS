# **Pendahuluan**

Dataset yang digunakan dalam analisis ini adalah “Tic-Tac-Toe Endgame” yang bersumber dari UCI Machine Learning Repository. Dataset ini berisi semua kemungkinan konfigurasi papan permainan Tic-Tac-Toe di akhir permainan, di mana 'x' diasumsikan sebagai pemain pertama. Tujuan dari dataset ini adalah untuk keperluan klasifikasi biner, yaitu menentukan apakah suatu konfigurasi papan merupakan kemenangan untuk pemain 'x'.

Dataset ini memiliki 958 data (instance) dan 10 atribut yang semuanya bersifat kategorikal. Dari 10 atribut tersebut, 9 atribut berfungsi sebagai fitur (merepresentasikan 9 kotak di papan) dan 1 atribut berfungsi sebagai kelas target (menunjukkan hasil akhir permainan).

# **Preprocessing (Encoding)**

Setelah kita memahami data kita melalui eksplorasi dan visualisasi, langkah selanjutnya adalah mempersiapkan data untuk model machine learning. Model machine learning tidak bisa memproses data teks ('x', 'o', 'positive', dll.), sehingga kita harus mengubahnya menjadi format angka. Proses ini disebut Encoding.

**Untuk dataset ini, kita akan menggunakan dua jenis encoding:**

- **Untuk Fitur (X)**: Kita akan menggunakan OneHotEncoder. Ini adalah pilihan terbaik karena fitur-fitur kita (posisi papan) bersifat nominal, artinya tidak ada urutan atau tingkatan ('x' tidak lebih baik dari 'o'). OneHotEncoder akan membuat kolom-kolom biner baru untuk setiap kemungkinan nilai (misalnya top-left-square_x, top-left-square_o, top-left-square_b).

- **Untuk Target (y)**: Kita akan menggunakan LabelEncoder. Ini sudah cukup karena kita hanya perlu mengubah dua kategori ('positive' dan 'negative') menjadi nilai numerik seperti 1 dan 0.

```python
# 1. Import semua library yang akan kita butuhkan
import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# 2. Muat dan siapkan data
tic_tac_toe_endgame = fetch_ucirepo(id=101)
df = pd.concat([tic_tac_toe_endgame.data.features, tic_tac_toe_endgame.data.targets], axis=1)

X = df.drop('class', axis=1)
y = df['class']

# 3. Preprocessing (Encoding)
encoder_X = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder_X.fit_transform(X)
X_encoded_df = pd.DataFrame(X_encoded, columns=encoder_X.get_feature_names_out(X.columns))

encoder_y = LabelEncoder()
y_encoded = encoder_y.fit_transform(y)

# 4. Pembagian Data
# Data dibagi menjadi 80% latih dan 20% uji
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded_df, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("--- Persiapan Data Selesai ---")
print(f"Data Latih: {X_train.shape[0]} baris")
print(f"Data Uji: {X_test.shape[0]} baris")
print("Data siap digunakan untuk melatih 3 model.")
```

```python
# Import library untuk encoding
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import pandas as pd

# DataFrame 'df' sudah ada dari langkah-langkah sebelumnya

print("--- Memulai Proses Preprocessing Data (Encoding) ---")

# Memisahkan antara fitur (X) dan target (y)
X = df.drop('class', axis=1)
y = df['class']

# 1. Encoding Fitur (X) dengan OneHotEncoder
# sparse_output=False agar hasilnya berupa array numpy biasa, bukan sparse matrix
encoder_X = OneHotEncoder(sparse_output=False)
X_encoded_array = encoder_X.fit_transform(X)

# Mengubah hasil encoding menjadi DataFrame baru dengan nama kolom yang jelas
X_encoded_df = pd.DataFrame(X_encoded_array, columns=encoder_X.get_feature_names_out(X.columns))

print("\n--- Fitur (X) Setelah di-OneHotEncoder ---")
print(f"Bentuk data fitur baru (jumlah baris, jumlah kolom): {X_encoded_df.shape}")
print("5 baris pertama dari data fitur yang sudah di-encode:")
print(X_encoded_df.head())

# 2. Encoding Target (y) dengan LabelEncoder
encoder_y = LabelEncoder()
y_encoded = encoder_y.fit_transform(y)

print("\n--- Target (y) Setelah di-LabelEncoder ---")
# Menampilkan mapping dari teks ke angka ('negative' -> 0, 'positive' -> 1)
print(f"Mapping kelas target: {dict(zip(encoder_y.classes_, encoder_y.transform(encoder_y.classes_)))}")
print("Contoh 5 data target pertama yang sudah di-encode:")
print(y_encoded[:5])
```

# **Klasifikasi**

**Alur kerja lengkap untuk tahap klasifikasi:**

- **Membagi Data:**Kita akan membagi dataset menjadi dua: 80% untuk melatih model (data latih) dan 20% untuk menguji model (data uji).
- **Membuat & Melatih Model:** Kita akan membuat model menggunakan algoritma Decision Tree Classifier, sama seperti pada contoh acuan Anda. Model ini akan "belajar" dari data latih.
- **Prediksi & Evaluasi:** Setelah dilatih, model akan diminta untuk menebak hasil dari data uji. Kemudian, kita akan membandingkan tebakan model dengan jawaban sebenarnya untuk mengukur seberapa baik kinerjanya.

## **1. Decision Tree Classifier**

```python

print("--- Melatih dan Mengevaluasi Model 1: Decision Tree ---")

# 1. Membuat dan melatih model
dt_model = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_model.fit(X_train, y_train)

# 2. Melakukan prediksi dan evaluasi
y_pred_dt = dt_model.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
report_dt = classification_report(y_test, y_pred_dt, target_names=encoder_y.classes_)

# 3. Menampilkan hasil
print(f"\nAkurasi Decision Tree: {accuracy_dt * 100:.2f}%\n")
print("Laporan Klasifikasi Decision Tree:")
print(report_dt)
```

**Penjelasan Hasil Akhir**

- Akurasi: Anda akan melihat akurasi model dalam bentuk persentase. Untuk dataset ini, model Decision Tree biasanya mencapai akurasi yang sangat tinggi.
- aporan Klasifikasi: Laporan ini akan memberikan rincian tentang seberapa baik model memprediksi kelas positive dan negative.

## **2. Naive Bayes Classifier**

```python
print("--- Melatih dan Mengevaluasi Model 2: Naive Bayes ---")

# 1. Membuat dan melatih model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# 2. Melakukan prediksi dan evaluasi
y_pred_nb = nb_model.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
report_nb = classification_report(y_test, y_pred_nb, target_names=encoder_y.classes_)

# 3. Menampilkan hasil
print(f"\nAkurasi Naive Bayes: {accuracy_nb * 100:.2f}%\n")
print("Laporan Klasifikasi Naive Bayes:")
print(report_nb)
```

**Penjelasan Hasil:**

Anda mungkin akan melihat bahwa akurasi Naive Bayes sedikit lebih rendah dibandingkan Decision Tree. Hal ini wajar, karena asumsi "naif"-nya (bahwa setiap posisi kotak di papan tidak saling memengaruhi) sebenarnya tidak sepenuhnya benar dalam permainan Tic-Tac-Toe. Namun, model ini tetap memberikan performa yang cukup baik dan proses pelatihannya sangat cepat.

## **3. K-Nearest Neighbors (KNN)**

```python
print("--- Melatih dan Mengevaluasi Model 3: K-Nearest Neighbors (KNN) ---")

# 1. Membuat dan melatih model
# n_neighbors=5 berarti kita akan melihat 5 tetangga terdekat
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# 2. Melakukan prediksi dan evaluasi
y_pred_knn = knn_model.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
report_knn = classification_report(y_test, y_pred_knn, target_names=encoder_y.classes_)

# 3. Menampilkan hasil
print(f"\nAkurasi K-Nearest Neighbors (k=5): {accuracy_knn * 100:.2f}%\n")
print("Laporan Klasifikasi K-Nearest Neighbors:")
print(report_knn)
```
